```python
from bs4 import BeautifulSoup 
import requests 
import csv
import pandas as pd
```

```python
req = requests.get("https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1")  # URL of the website which you want to scrape
content = req.content # Get the content
```

```python
soup = BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
desc = soup.find_all('div' , class_='_4rR01T')
desc
```

```python
descriptions = [] # Create a list to store the descriptions
for i in range(len(desc)):
    descriptions.append(desc[i].text)
len(descriptions)
```

```python
descriptions
```

```python
cclass=soup.find_all('li' , class_='rgWa7D')
cclass
```

```python
commonclass= [] # Create a list to store the descriptions
for i in range(len(cclass)):
    commonclass.append(cclass[i].text)
len(commonclass)
```

```python
commonclass
```

```python
# Create empty lists for the features
processors=[]
ram=[]
os=[]
storage=[]
inches=[]
warranty=[]
for i in range(0,len(commonclass)):
    p=commonclass[i]
    if("Core" in p): 
        processors.append(p)
    elif("RAM" in p): 
        ram.append(p)
# If RAM is present in the text then append it to the ram list. Similarly do this for the other features as well
    elif("HDD" in p or "SSD" in p):
        storage.append(p)
    elif("Operating" in p):
        os.append(p)
    elif("Display" in p):
        inches.append(p)
    elif("Warranty" in p):
        warranty.append(p)
```

```python
print(len(processors))
print(len(warranty))
print(len(os))
print(len(ram))
print(len(inches))
```

```python
price = soup.find_all('div', class_="_30jeq3 _1_WHN1") 
# Extracting price of each laptop from the website
prices = []
for i in range(len(price)):
    prices.append(price[i].text)
len(prices)
```

```python
prices
```

```python
rating = soup.find_all('div',class_='_3LWZlK') 
#Extracting the ratings of each laptop from the website
ratings = []
for i in range(len(rating)):
    ratings.append(rating[i].text)
len(ratings)
```

```python
ratings
```

```python
df = {'Description':descriptions,'Processor':processors,'RAM':ram,'Operating System':os,'Storage':storage,'Display':inches,'Warranty':warranty,'Price':prices}
dataset = pd.DataFrame.from_dict(df, orient='index')
dataset = dataset.transpose()
```
